### Group example

````python
from ecc import FieldElement, Point

prime = 223
a = FieldElement(0, prime)
b = FieldElement(7, prime)
G = Point(FieldElement(47, prime), FieldElement(71, prime), a, b)
inf = Point(None, None, a, b)
total = G
count = 1

while total != inf:
    print('{}:{}'.format(count, total))
    total += G
    count += 1
print('{}:{}'.format(count, total))
````
